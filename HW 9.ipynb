```python
from random import random
from numpy import arange
from pylab import plot,xlabel,ylabel,show


NBi = 10000             
NPb = 0 
NTl = 0
Bi = 0
tau1 = 46*60
tau2 = 2.2*60
tau3 = 3.3*60
h = 1.0  
pt1 = 1-2**(-h/tau1) 
pt2 = 1-2**(-h/tau2)
pt3 = 1-2**(-h/tau3)
tmax = 20000

tpoints = arange(0.0,tmax,h) 
NBipoints = [] 
NTlpoints = []
NPbpoints = []
Bipoints = []

for t in tpoints:
    NBipoints.append(NBi)
    NTlpoints.append(NTl)
    
    decay = 0
    for i in range(NBi): 
         if random() < pt1:
            decay = 1
    NBi -= decay  
    NTl += decay
for t in tpoints:
    NBipoints.append(NBi)
    NPbpoints.append(NPb)
    
    decay = 0
    for i in range(NBi):
        if random() < pt1:
            decay = 1
        NBi -= decay
        NPb += decay
for t in tpoints:
    NTlpoints.append(NTl)
    NPbpoints.append(NPb)
    
    decay = 0
    for i in range(NTl):
        if random() < pt2:
            decay = 1
        NTl -= decay
        NPb += decay
for t in tpoints: 
    NPbpoints.append(NPb)
    Bipoints.append(Bi)
    
    decay = 0
    for i in range(NPb):
        if random() < pt3:
            decay = 1
        NPb -= decay
        NPb += decay
plot(tpoints, NBipoints, c='m') 
plot(tpoints, NTlpoints, c='g') 
plot(tpoints, NPbpoints, c='r')
plot(tpoints, Bipoints, c='b')
xlabel("Time")
ylabel("Number of atoms")
show()
```
